```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 2 * np.pi * 4, 1e-2)
x1 = np.arange(0, 2 * np.pi, (2 * np.pi) / 7)
x2 = np.arange(0, 2 * np.pi * 4, (2 * np.pi * 4) / 25)
s = np.sin(x)
sin1 = np.sin(x1)
sin2 = np.sin(x2)
plt.plot(x, s)
plt.plot(x1, sin1, "*")
plt.plot(x2, sin2, ".")
```
